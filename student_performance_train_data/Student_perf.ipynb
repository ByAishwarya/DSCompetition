```python
# ÙÙ‚Ø· Ø³ØªÙˆÙ†â€ŒÙ‡Ø§ÛŒ Ø¹Ø¯Ø¯ÛŒ
numeric_cols = df.select_dtypes(include=['number'])

# Ù…Ø­Ø§Ø³Ø¨Ù‡ Ù‡Ù…Ø¨Ø³ØªÚ¯ÛŒ
corr_matrix = numeric_cols.corr()

# Ø¯ÛŒØ¯Ù† Ù‡Ù…Ø¨Ø³ØªÚ¯ÛŒ Ù‡Ø± Ù…ØªØºÛŒØ± Ø¨Ø§ Ù†Ù…Ø±Ù‡ Ù†Ù‡Ø§ÛŒÛŒ
corr_with_G3 = corr_matrix['G3'].sort_values(ascending=False)
corr_with_G3
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x=corr_with_G3.values, y=corr_with_G3.index, palette='viridis')
plt.title("Correlation of Features with Final Grade (G3)")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Feature")
plt.show()
```

```python
import scipy.stats as stats

categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    groups = [group["G3"].values for name, group in df.groupby(col)]
    if len(groups) > 1:
        f, p = stats.f_oneway(*groups)
        print(f"{col}: p-value = {p:.4f}")
```

```python
import matplotlib.pyplot as plt

# Ø¯Ø§Ø¯Ù‡â€ŒÙ‡Ø§
labels = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
          'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
          'higher', 'internet', 'romantic']
p_values = [0.2928, 0.0882, 0.1172, 0.2612, 0.0622, 0.0193, 0.5161, 0.2511,
            0.3388, 0.2630, 0.5059, 0.0118, 0.7573, 0.2918, 0.0001, 0.0818, 0.0152]

# Ø±Ù†Ú¯â€ŒÙ‡Ø§ Ø¨Ø± Ø§Ø³Ø§Ø³ Ù…Ø¹Ù†Ø§Ø¯Ø§Ø±ÛŒ Ø¢Ù…Ø§Ø±ÛŒ
colors = ['#d62728' if p < 0.05 else '#1f77b4' for p in p_values]

# Ø±Ø³Ù… Ù†Ù…ÙˆØ¯Ø§Ø±
plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(labels, p_values, color=colors)
ax.axvline(0.05, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('p-value')
ax.set_title('ANOVA p-values for Categorical Variables')
plt.tight_layout()
plt.show()
```
